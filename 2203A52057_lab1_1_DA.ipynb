## ***scatter plot***

```python

features = ['age', 'bmi', 'children','sex','smoker','region']

for feature in features:
    plt.figure(figsize=(8, 5))
    sns.scatterplot(data=df, x=feature, y='expenses', hue='smoker', style='sex', s=100)
    plt.title(f'Scatter Plot: {feature} vs Expenses')
    plt.xlabel(feature)
    plt.ylabel('Expenses')
    plt.legend(title='Smoker Status')
    plt.show()
    print()
```
